# Output Layer

출력층(Output Layer)은 신경망의 마지막 층이다.

출력층은 모든 문제에서 똑같지 않다.

- 회귀: 연속형 수치를 예측
- 이진분류: 두 클래스 중 하나를 예측
- 다중분류: 여러 클래스 중 하나를 예측

즉, 어떤 문제를 풀고 있는지에 따라 마지막 출력 노드의 개수와 출력값의 해석 방식이 달라진다.

## 문제 유형별 출력층 비교

| 문제 유형 | 출력 노드 수 | 마지막 활성화 함수 | 출력값 해석 |
|---|---:|---|---|
| 회귀 | 1개(또는 여러 개) | 보통 없음 | 연속형 수치 |
| 이진분류 | 1개 | sigmoid 사용 가능 | 양성 클래스일 확률처럼 해석 가능 |
| 다중분류 | 클래스 수만큼 | softmax 사용 가능 | 각 클래스 확률처럼 해석 가능 |

정리하면,
출력층은 단순히 마지막에 값 하나를 내는 층이 아니라
문제 유형에 맞는 형태로 최종 예측값을 만들어내는 층이라고 볼 수 있다.

```python
import torch
import matplotlib.pyplot as plt
```

## 1. 회귀 출력층

회귀 문제에서는 연속형 값을 예측한다.
예를 들어 집값, 온도, 매출처럼 숫자 자체가 정답인 경우가 여기에 해당한다.

이 경우 마지막 출력층에서는 보통 값을 그대로 사용한다.
즉, 별도의 활성화 함수를 두지 않고 마지막 선형 출력값을 그대로 예측값으로 해석하는 경우가 많다.

```python
# 입력 4개 샘플이 있었을 경우 연속형 수치 출력 값
regression_output = torch.tensor([[210.5], [315.2], [178.9], [402.7]])

print('shape : ', regression_output.shape)
```

## 2. 이진분류 출력층

이진분류는 두 클래스 중 하나를 예측하는 문제이다.

이진분류에서는 마지막 출력 노드가 1개인 경우가 많다.

이때 마지막 값은 바로 0 또는 1이 아니라,먼저 logit이라는 점수 형태로 나오는 경우가 많다.

### logit, probability, threshold

1. logit  
   - sigmoid를 통과시키기 전의 점수
   - 아직 확률은 아니다

2. probability  
   - sigmoid를 통과한 뒤의 값
   - 0과 1 사이 값이며, 양성 클래스일 확률처럼 해석할 수 있다

3. prediction  
   - probability에 threshold를 적용해 최종적으로 0 또는 1로 판단한 결과

```python
# 모델의 마지막 선형 출력 값을 가정

logits = torch.tensor([[-2.0], [-0.3], [0.2], [1.5], [3.0]])

probabilities = torch.sigmoid(logits)

print('logits')
print(logits)

print('probabilities')
print(probabilities)
```

### threshold가 필요한 이유

sigmoid를 거친 값은 확률처럼 해석할 수 있지만,
모델이 최종적으로 어느 클래스로 판단할지를 정하려면 기준이 필요하다.

그 기준이 threshold이다.

보통은 0.5를 많이 사용하지만, 상황에 따라 0.3, 0.7처럼 다르게 잡을 수도 있다.

```python
for threshold in [0.3, 0.5, 0.7]:
    predictions = (probabilities >= threshold).int()
    print(f'threshold : {threshold}')
    print(predictions.squeeze().tolist())
    print()
```

## 3. 다중분류 출력층

다중분류는 여러 클래스 중 하나를 예측하는 문제이다.

이 경우 마지막 출력 노드는 클래스 수만큼 필요하다. 그리고 각 클래스에 대한 점수(logits)가 출력된다.

### logits, softmax, argmax

다중분류에서도 먼저 logits가 나온다.

예를 들어 클래스가 3개라면, 출력층은 다음처럼 3개의 점수를 만들 수 있다.

- class 0 점수
- class 1 점수
- class 2 점수

이 점수들에 softmax를 적용하면, 각 클래스일 확률처럼 해석할 수 있는 값으로 바뀐다.

마지막으로 argmax를 적용하면, 가장 큰 값을 가진 클래스 번호를 최종 예측으로 선택할 수 있다.

```python
# 샘플 3개, 클래스 4개 가정
multi_logits = torch.tensor([
    [2.0, -1.5, -1.0, 0.3],
    [1.0, -0.7, -2.0, 0.5],
    [-3.0, -1.2, 3.0, 0.1],
    ])

# 각 행(샘플) 안에서 클래스 방향으로 계산
multi_probabilities = torch.softmax(multi_logits, dim=1)

# 각 샘플 행 안에서 가장 큰 값의 위치를 고른다
multi_predictions = torch.argmax(multi_probabilities, dim=1)

print('multi_logits')
print(multi_logits)

print('multi_probabilities')
print(multi_probabilities)

print('multi_predictions')
print(multi_predictions)
```

## 4. 출력층과 손실함수

출력층은 손실함수와도 연결되어 설계된다.

| 문제 유형 | 출력층 해석 | 자주 연결되는 손실함수 |
|---|---|---|
| 회귀 | 연속형 수치 | MSELoss |
| 이진분류 | sigmoid 확률 또는 logits | BCELoss, BCEWithLogitsLoss |
| 다중분류 | softmax 확률 또는 logits | CrossEntropyLoss |

주의할 점은, 실전에서는 활성화 함수를 직접 적용하기보다 logits를 그대로 넣는 손실함수를 쓰는 경우도 많다는 점이다.

예를 들어,
- 이진분류에서는 `BCEWithLogitsLoss`
- 다중분류에서는 `CrossEntropyLoss`

를 자주 사용한다.
